# Part 3 :
##  Point processing techniques

```python
from skimage.color import rgb2lab , lab2rgb
```

```python
img_p=img[:,:, :3]
img_pp = (img_p * 255).astype(np.uint8)
plt.imshow(img_pp)
```

```python
img_pp.shape
```

```python
darkened_img = np.clip(img_pp * 0.5, 0, 255).astype(np.uint8)

plt.imshow(darkened_img)
```

```python
lab_img = rgb2lab(img_pp)

# Split the LAB image into L, A, and B channels
l_channel, a_channel, b_channel = lab_img[:, :, 0], lab_img[:, :, 1], lab_img[:, :, 2]

# Darken the image by scaling the L channel
# Multiply the L channel by a factor (e.g., 0.5) to darken the image
darkened_l_channel = np.clip(l_channel * 0.5, -128, 127)  # LAB L channel range is [-128, 127]

# Merge the darkened L channel with the original A and B channels
darkened_lab_img = np.stack([darkened_l_channel, a_channel, b_channel], axis=-1)

# Convert LAB back to RGB
darkened_img = lab2rgb(darkened_lab_img)

# Display the original and darkened images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(img)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(darkened_img)
axes[1].set_title('Darkened Image')
axes[1].axis('off')

plt.show()
```
